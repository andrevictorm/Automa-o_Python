```python
print(f' {e},\n {nome},\n {tel},\n {area},\n {cargo},\n {sen},\n {lnkd}')
```

```python
tab=[]
for i in range (len(e)):
    test={}
    test['Nome']=nome[i]
    test['Email']=e[i]
    test['Tel']=tel[i]
    test['Area']=area[i]
    test['Cargo']=cargo[i]
    test['Senioridade']=sen[i]
    test['Preferencia']=pref[i]
    test['LinkedIn']=lnkd[i]
    test['Empresa']=empresa[i]
    tab.append(test)
print(tab)
```
